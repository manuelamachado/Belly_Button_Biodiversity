```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func, desc
```

```python
engine = create_engine('sqlite:///DataSets/belly_button_biodiversity.sqlite')

Base = automap_base()
Base.prepare(engine,reflect = True)
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
Samples = Base.classes.samples
Metadata = Base.classes.samples_metadata
Otu = Base.classes.otu
```

```python
session = Session(bind = engine)
```

* /names

```python
def names():
    sample_names = []
    name_result = session.query(Metadata.SAMPLEID)

    for result in name_result:
        sample_names.append('BB_' + str(result[0]))

    return(sample_names[0])

names()
```

* /otu

```python
def otu():
    otu_description = []
    otu_result = session.query(Otu.lowest_taxonomic_unit_found)
    for result in otu_result:
        otu_description.append(result[0])
        
    return(otu_description[0])

otu()
```
